```python
# function definition of quiz
def quiz(user_ans,actual_ans):
    attempt=3
    # to get the access of score variable through out the program
    global score
    ans=True
    while ans and attempt>0:
        # to convert into upper case in order to avoid case sensitive problems and to ignore space issues
        cap_ans=user_ans.upper().replace(' ','')
        if cap_ans==actual_ans:
            print("Wow...Excellent!!! Your guess is absolutely correct.")
            score+=1
            print('Score Status: ',score,'\n')
            ans=False
        else:
            attempt-=1
            if attempt >0:
                print('Oops!!!, It is a wrong guess.\nYou have',attempt,'attempt left.\n')
                user_ans=input('Try again: ')
        if attempt==0:
            print('Sorry, no more attempts left.\nThe correct answer is',actual_ans,'\n')
score=0

print('Guess the animal!!!\n')

q1=input('Q1. Which is the fastest land animal? ')
quiz(q1,'CHEETAH')

q2=input('Q2. Which is the tallest animal? ')
quiz(q2,'GIRAFFE')

q3=input('Q3. Which is the largest mammal in the world? ')
quiz(q3,'BLUEWHALE')

q4=input('Q4. Which is the only land animal that cannot jump? ')
quiz(q4,'ELEPHANT')

q5=input('Q5. Which is the only mammal that can fly? ')
quiz(q5,'BAT')

print('**********Congratulations!!!**********\nYou have successfully completed the quiz.\nThank you for your participation.\n Your final score: ',score,'out of 5.')
```
